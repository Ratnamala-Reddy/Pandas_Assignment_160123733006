```python
import pandas as pd
import numpy as np
```

1st Question

```python
df =pd.read_csv("/content/Titanic-Dataset.csv")
df.head()
```

```python
#Create Age Series
age_series = df['Age']
```

```python
#Descriptive Statistics
print(age_series.describe())
```

```python
# Create Age Groups
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
# Fatality Rate
fatality = df.groupby('AgeGroup')['Survived'].apply(lambda x: 1 - x.mean())
print(fatality)
```

**Conclusion** : Most passengers were young adults, and this group had the highest fatality rate. This happened mainly because children and women were given priority during rescue, leaving many adult males behind.

3rd Question

```python
#Find Missing Values
df.isnull().sum()
```

```python
#Fill with Mean and Median
mean_age = df['Age'].fillna(df['Age'].mean())
median_age = df['Age'].fillna(df['Age'].median())
print(df['Age'].mean(), df['Age'].median())
```

**Conclusion** : The Age column has missing values, and using the median is better than the mean. This is because median is not affected much by extreme values and gives a more realistic estimate.

5th Question


```python
#Create FamilySize and  SurvivalScore
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['SurvivalScore'] = np.log1p(df['Fare']) * df['FamilySize']
```

```python
# Sort
df_sorted = df.sort_values(by='SurvivalScore', ascending=False)
df_sorted.head()
```

**Conclusion** : By creating FamilySize and combining it with Fare, we get a SurvivalScore that gives a better idea of survival chances. This shows how creating new features can improve analysis.

8th Question

```python
# Function
def categorize(age):
    if age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'
df['AgeCategory'] = df['Age'].apply(categorize)
```

```python
#pivot
pivot = pd.pivot_table(df, values='Survived',
                       index='AgeCategory',
                       columns='Sex',
                       aggfunc='mean')
print(pivot)
```

**Conclusion** : Dividing passengers into age categories like Child, Adult, and Senior makes the analysis clearer. It shows that children and females had better survival chances.
